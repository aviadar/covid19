```python
sys.path.insert(0,'/content/proj')

from utilities.sentence_utility import SentenceUtil
```

```python
messages = [
        # Smartphones
        "I like my phone",
        "My phone is not good.",
        "Your cellphone looks great.",

        # Weather
        "Will it snow tomorrow?",
        "Recently a lot of hurricanes have hit the US",
        "Global warming is real",

        # Food and health
        "An apple a day, keeps the doctors away",
        "Eating strawberries is healthy",
        "Is paleo better than keto?",

        # Asking about age
        "How old are you?",
        "what is your age?",
    ]

sentence_util = SentenceUtil(messages)
k = 3
compared_index = 2
topk_ind, topk_sim = sentence_util.get_k_most_similar(compared_index=compared_index, k=k)
```

```python

print('top '+ str(k) +' similarities to: ' + messages[compared_index])
for i in range(k):
    print(messages[topk_ind[i]], topk_sim[i])
```
