```python
import pandas as pd
import json
import tqdm
```

note that in order to run this notebook, you need to download and extract the [covid-19 dataset](https://www.kaggle.com/datasets/allen-institute-for-ai/CORD-19-research-challenge?resource=download)

```python
# path of the downloaded kaggle archive
DIR_PATH = r'C:\Users\aviadar\Downloads\archive/'
```

```python
df = pd.read_csv(DIR_PATH + 'metadata.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.dropna(subset=['pdf_json_files', 'title', 'abstract'], inplace=True)
df.shape
```

```python
df.sort_values(by=['publish_time'], ascending=False, inplace=True)
df.reset_index(inplace = True)
```

```python
df.head()
```

```python
def get_text(filename):
    with open(DIR_PATH + filename) as file:
        json_content = json.load(file)
    
    body_text = []
    for txt in json_content['body_text']:
        body_text.append(txt['text'])

    body_text = '\n'.join(body_text)
    return body_text
```

```python
article_num = 20000
covid_df = pd.DataFrame(columns=["title", "cord_uid", "abstract", "body"])
with tqdm.tqdm(total=article_num) as pbar:
    for _, row in df.iterrows():
        row_df = pd.DataFrame(columns=["title", "cord_uid", "abstract", "body"])
        if type(row["pdf_json_files"]) == str:
            json_filepath = row["pdf_json_files"]
        else:
            continue
        if ";" in json_filepath:
            continue
        body = get_text(json_filepath)

        row_df['title'] = [row['title']]
        row_df['cord_uid'] = [row['cord_uid']]
        row_df['abstract'] = row['abstract']
        row_df['body'] = [body]

        covid_df = pd.concat([covid_df, row_df], ignore_index=True)
        pbar.update(1)
        
        if covid_df.shape[0] >= article_num:
            break
pbar.close()    
```
