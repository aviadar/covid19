```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
drive.mount('/content/drive')
```

# Data Loading

Loading the preprocessed data (for more info, go to: data_preparation.ipynb)

```python
df = pd.read_csv('/content/drive/MyDrive/AdvancedML/HW3/covid_df.csv')
df
```

# WordCloud

```python
titles = "".join(df.title)  
# Plot the wordcloud
f, ax = plt.subplots(figsize=(16, 8))
plt.imshow(WordCloud(width=1000, height=400, stopwords=STOPWORDS, background_color='white').generate(titles), interpolation="bilinear")
plt.axis('off')
plt.title('Title Wordcloud')
```

```python
abstracts = "".join(df.abstract)  
# Plot the wordcloud
f, ax = plt.subplots(figsize=(16, 8))
plt.imshow(WordCloud(width=1000, height=400, stopwords=STOPWORDS, background_color='white').generate(abstracts), interpolation="bilinear")
plt.axis('off')
plt.title('Abstract Wordcloud')
```

```python
bodies = "".join(df.body)  
# Plot the wordcloud
f, ax = plt.subplots(figsize=(16, 8))
plt.imshow(WordCloud(width=1000, height=400, stopwords=STOPWORDS, background_color='white').generate(bodies), interpolation="bilinear")
plt.axis('off')
plt.title('Bodies Wordcloud')
```
