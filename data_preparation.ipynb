```python
def get_text(filename):
#     with zipfile.ZipFile(r'C:\Users\aviadar\Downloads\archive.zip', 'r') as zipObj:
#     # Get a list of all archived file names from the zip
#         listOfFileNames = zipObj.namelist()    
#         # Iterate over the file names
#         for fileName in listOfFileNames:
#         # Check filename endswith csv
#             if fileName.endswith(filename):
#                 # Extract a single file from zip
#                 zipObj.extract(fileName, 'temp_csv')
#                 break

    with open(r'C:\Users\aviadar\Documents\AdvancedML\HW3/'+filename) as file:
        json_content = json.load(file)
    
    body_text = []
    for txt in json_content['body_text']:
        body_text.append(txt['text'])

    body_text = '\n'.join(body_text)
    return body_text
```

```python
article_num = 3
covid_df = pd.DataFrame(columns=["title", "cord_uid", "abstract", "body"])
with tqdm.tqdm(total=article_num) as pbar:
    for _, row in df.iterrows():
        row_df = pd.DataFrame(columns=["title", "cord_uid", "abstract", "body"])
        if type(row["pdf_json_files"]) == str:
            json_filepath = row["pdf_json_files"]
        else:
            continue
        if ";" in json_filepath:
            continue
        body = get_text(json_filepath)

        row_df['title'] = [row['title']]
        row_df['cord_uid'] = [row['cord_uid']]
        row_df['abstract'] = row['abstract']
        row_df['body'] = [body]

        covid_df = pd.concat([covid_df, row_df], ignore_index=True)
        pbar.update(1)
        
        if covid_df.shape[0] >= article_num:
            break
pbar.close()    
```
